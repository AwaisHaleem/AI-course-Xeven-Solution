```python
# import necessary libraries
import pandas as pd

# import titanic dataset
df = pd.read_csv("titanic_dataset.csv")
```

```python
# Identify the data types of each feature.
df.dtypes
```

```python
# Explore categorical features by counting unique values in each category.

# As we explored earliar categorical features
cat_features = ["Survived", "Sex", "Pclass", "Embarked"]


for feature in cat_features:                                # loop throght categorical features lis
    unique_values = df[feature].unique()                    # get unigue values in the feature
    unique_values_count = df[feature].value_counts()        # counts for every unique value in the feature
    print(f"{feature} feature has {unique_values} values\n")
    print(unique_values_count, "\n\n")
```

```python
# Investigate missing values in the dataset and determine the percentage of missing values for each feature.

for feature in df.columns:                                              # loop through all features
    missing_values_count = df[feature].isnull().sum()                   # get missing value count for the feature
    missing_values_percentage = (missing_values_count / len(df)) * 100  # calculate percentage of missing values
    print(f"{feature}:  {missing_values_count} missing values:   {missing_values_percentage:.2f}%")
```
