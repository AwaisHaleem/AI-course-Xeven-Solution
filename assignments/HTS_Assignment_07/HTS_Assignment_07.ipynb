```python
# let's convert ocean proximity values in category values
df["ocean_proximity"] = df["ocean_proximity"].astype("category")
df.dtypes
```

* ocean_proximity has been changer from object to category

```python
# change ocean_proximity values in numbers
ocean_proximity = pd.get_dummies(df["ocean_proximity"], drop_first=False)
ocean_proximity.head()
```

```python
# join the data frames
df = pd.concat([df, ocean_proximity], axis=1)
df.head()
```

```python
# rename the <1h ocean
df.rename({"<1H OCEAN": "1H_OCEAN"},axis=1,  inplace=True)
df.head()
```

```python
# now drops ocean proximity column
df.drop("ocean_proximity", axis=1, inplace=True)
df.columns
```

```python
# let's plot the corraltion of df
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), annot=True, cmap='RdBu', fmt="0.2g")
plt.axhline(y=8, color="purple", linewidth=4)
plt.axhline(y=9, color="purple", linewidth=4)
plt.show()
```

<h1>Observation</h1>
- we will choose house value as our target label<br>
- we will drop all features whose relation in less than 0.1<br>
- we only have feautes whose relation value is equal to or greater than 0.1<br>
- we will drop longitude, total_bedrooms, population, housholds, ISLAND feature

```python
# let's drop longitude, total_bedrooms, population, housholds, ISLAND feature
df.drop(["longitude", "total_bedrooms", "population", "households", "ISLAND"], axis=1, inplace=True)
df.columns
```
