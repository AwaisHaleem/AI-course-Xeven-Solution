Sentiment analysis

```python
# function for sentiment analysis

def analyse_sentiment(text):
    analyzer = transformers.pipeline("sentiment-analysis") # importing sentiment analysis pipeline
    return analyzer(text)
```

```python
analyse_sentiment("absolutely loved the movie! The acting was phenomenal and the plot kept me engaged throughout.")
```

```python
analyse_sentiment("The customer service was terrible. I had to wait for hours and no one seemed to care about my issue.")
```

sentiment analysis with FinancialBERT-Sentiment-Analysis model

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline

model = BertForSequenceClassification.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis",num_labels=3)
tokenizer = BertTokenizer.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis")

nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
sentences = [
    "The novel had a slow start, but the ending was absolutely breathtaking.",
    "I can't stop laughing at that hilarious comedy show. My cheeks hurt!",
    "The restaurant was disappointing. The food was cold, and the service was slow.",
    "The weather is beautiful today. I'm feeling so happy and content."
]

result = nlp(sentences)
result
```

```python
!pip install ultralyticsplus
from ultralyticsplus import YOLO, render_result

# load model
model = YOLO('foduucom/thermal-image-object-detection')

# set model parameters
model.overrides['conf'] = 0.25  # NMS confidence threshold
model.overrides['iou'] = 0.45  # NMS IoU threshold
model.overrides['agnostic_nms'] = False  # NMS class-agnostic
model.overrides['max_det'] = 1000  # maximum number of detections per image

image = "/content/sample_data/img.jpg"

# perform inference
results = model.predict(image)
```

```python
# observe results
print(results[0].boxes)
render = render_result(model=model, image=image, result=results[0])
render.show()
```
